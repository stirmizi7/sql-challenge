```python
# Assuming you have a DataFrame 'dept_emp_data' or 'departments_data' with department information
sales_and_dev = pd.merge(employee_data, dept_emp_data, on='emp_no')
sales_and_dev = pd.merge(sales_and_dev, departments_data, on='dept_no')
sales_and_dev = sales_and_dev[sales_and_dev['dept_name'].isin(['Sales', 'Development'])]
sales_and_dev[['emp_no', 'last_name', 'first_name', 'dept_name']]
```

```python
last_name_counts = employee_data['last_name'].value_counts().reset_index()
last_name_counts.columns = ['last_name', 'count']
last_name_counts = last_name_counts.sort_values(by='count', ascending=False)
```

```python
import matplotlib.pyplot as plt

# Assuming last_name_counts contains the last name frequency counts
plt.figure(figsize=(10, 6))
plt.bar(last_name_counts['last_name'], last_name_counts['count'])
plt.xlabel('Last Name')
plt.ylabel('Count')
plt.title('Employee Last Name Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()
```
